Read in the data from the csv file above. Skip all the comment lines, create a header for the dataframe based on the fields list in the comment line that starts with **# Fields:**. Drop the first column `query acc.ver`. Set the `subject acc.ver` as the index of the dataframe. The last column of the dataframe should be `publications`. Show the first five rows of the resulting dataframe.

```python
import pandas as pd
import io
import requests
```

```python
url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/blast_results.csv"

data = pd.read_csv(url, comment='#', header=None)

fields = [
    "query acc.ver", "subject acc.ver", "% identity", "alignment length", "mismatches",
    "gap opens", "q. start", "q. end", "s. start", "s. end", "evalue", "bit score",
    "% positives", "bonds", "publications", "extra col"
]

data.columns = fields

data = data.drop(columns=["query acc.ver"]).set_index("subject acc.ver")
data = data.iloc[1:]
data = data.iloc[:, :-1]

data
```

```python
# prompt: Using dataframe data: Create a plot that shows the correlation between % identity and % positives.

import altair as alt

alt.Chart(data).mark_point().encode(
    x='% identity',
    y='% positives'
).interactive()
```

### Question  5
Add a column `Protein source` to the dataframe from Question 1 based on the data in this csv file: https://raw.githubusercontent.com/csbfx/advpy122-data/master/protein_source.csv. Merge the data when the `subject acc.ver` from the dataframe equals to `Protein` in the csv file. Entries without a matching protein will have `NA` as `Protein source`.

```python
protein_source_url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/protein_source.csv"

protein_source_data = pd.read_csv(protein_source_url, header=0)

protein_source_data = protein_source_data.rename(columns={'Source': 'Protein source'})

protein_source_data['Protein'] = pd.to_numeric(protein_source_data['Protein'], errors='coerce')

data.index = pd.to_numeric(data.index, errors='coerce')

data = pd.merge(data, protein_source_data, left_index=True, right_on='Protein', how='left')

data['Protein source'] = data['Protein source'].fillna('NA')

# Display the updated DataFrame
data.head()
```
